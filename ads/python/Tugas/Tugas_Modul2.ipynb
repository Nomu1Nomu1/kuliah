## Singly Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        
    def insert(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        
    def delete(self, key):
        temp = self.head
        
        if temp and temp.data == key:
            self.head = temp.next
            temp = None
            return
        
        prev = None
        while temp and temp.data != key:
            prev = temp
            temp = temp.next
            
        if not temp:
            return
        
        prev.next = temp.next
        temp = None
        
    def traverse(self):
        temp = self.head
        
        while temp:
            print(temp.data, end=' -> ')
            temp = temp.next
        print("None")
        
sll = SinglyLinkedList()
sll.insert(10)
sll.insert(20)
sll.insert(40)
sll.traverse()

sll.delete(20)
sll.traverse()
```

## Doubly Linked List

```python
class DNode:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
        
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        
    def insert(self, data):
        new_node = DNode(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp
        
    def delete(self, key):
        temp = self.head
        while temp and temp.data != key:
            temp = temp.next
        if not temp:
            return
        if temp.prev:
            temp.prev.next = temp.next
        else:
            self.head = temp.next
            
        if temp.next:
            temp.next.prev = temp.prev
            
        temp = None
        
    def traverse_forward(self):
        temp = self.head
        while temp:
            print(temp.data, end=" <-> ")
            last = temp
            temp = temp.next
        print("None")
    
    def traverse_backward(self):
        temp = self.head
        while temp and temp.next:
            temp = temp.next
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.prev
        print("None")

dll = DoublyLinkedList()
dll.insert(10)
dll.insert(20)
dll.insert(30)
dll.traverse_forward()

dll.delete(20)
dll.traverse_forward()
dll.traverse_backward()
```

## Circular Linked List

```python
class Cnode:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class CircularLinkedList:
    def __init__(self):
        self.head = None
        
    def insert(self, data):
        new_node = Cnode(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = new_node
        new_node.next = self.head
        
    def delete(self, key):
        if not self.head:
            return
        
        temp = self.head
        prev = None
        
        if temp.data == key:
            while temp.next != self.head:
                temp = temp.next
            if self.head == self.head.next:
                self.head = None
            else:
                temp.next = self.head.next
                self.head = self.head.next
            return
        
        temp = self.head
        while temp.next != self.head and temp.next.data != key:
            temp = temp.next
        if temp.next.data == key:
            temp = temp.next.next
            
    def traverse(self):
        if not self.head:
            print("List kosong")
            return
        temp = self.head
        while True:
            print(temp.data, end=" -> ")
            temp = temp.next
            if temp == self.head:
                break
        print("(back to head)")
        
cll = CircularLinkedList()
cll.insert(10)
cll.insert(100)
cll.insert(1000)
cll.traverse()

cll.delete(10)
cll.traverse()
```

## 5. 
Arah navigasi:
- Singly Linked List: Merupakan bentuk paling dasar. Setiap simpul hanya tahu "siapa selanjutnya", tetapi tidak tahu "siapa sebelumnya". Ini membuat penelusuran hanya bisa dilakukan dari awal (head) hingga akhir (tail).

- Doubly Linked List: Memberikan fleksibilitas lebih dengan adanya dua penunjuk di setiap simpul. Kita bisa bergerak maju ke simpul berikutnya atau mundur ke simpul sebelumnya dengan mudah. Hal ini sangat menyederhanakan beberapa operasi.

- Circular Linked List: Konsep utamanya adalah tidak ada akhir yang sesungguhnya (NULL pointer di akhir). Simpul terakhir terhubung kembali ke simpul pertama. Ini memungkinkan traversal yang berkelanjutan dan efisien untuk aplikasi tertentu. Circular Linked List bisa diimplementasikan sebagai Singly Circular Linked List (hanya maju) atau Doubly Circular Linked List (maju dan mundur).

Efisiensi Insert/Delete:
Efisiensi operasi diukur dalam notasi Big O, di mana O(1) berarti waktu konstan (sangat cepat) dan O(n) berarti waktu linear (kecepatan bergantung pada jumlah elemen, 'n').

- Operasi di Awal: Semua tipe linked list unggul dalam operasi di awal. Menambah atau menghapus simpul di posisi kepala hanya memerlukan beberapa penyesuaian penunjuk tanpa perlu iterasi.

- Operasi di Akhir: Di sinilah perbedaan signifikan terlihat. Singly Linked List harus menelusuri dari awal hingga akhir untuk menemukan simpul terakhir sebelum melakukan operasi, sehingga memiliki kompleksitas O(n). Sebaliknya, Doubly Linked List (dengan asumsi ia melacak simpul terakhir) dan Circular Linked List dapat mengakses simpul terakhir secara langsung dalam waktu O(1).

- Operasi di Tengah: Untuk menyisipkan atau menghapus simpul di tengah, semua tipe umumnya memerlukan penelusuran dari awal untuk menemukan posisi yang benar, sehingga kompleksitasnya adalah O(n). Namun, pada Doubly Linked List, setelah posisi ditemukan, proses modifikasi penunjuk lebih mudah karena memiliki akses langsung ke simpul sebelumnya.

Kegunaan:
- Jika membangun aplikasi sederhana yang hanya memerlukan penambahan dan penghapusan di satu ujung (seperti Stack) dan ingin menghemat memori, Singly Linked List merupakan pilihan yang tepat.

- Jika aplikasi memerlukan navigasi dua arah, seperti riwayat browser atau fungsionalitas undo, Doubly Linked List menjadi pilihan yang jauh lebih superior meskipun dengan biaya memori tambahan.

- Untuk sistem yang mengelola sumber daya secara bergiliran atau membutuhkan aliran data yang berkelanjutan, seperti penjadwalan CPU atau playlist lagu yang berulang, Circular Linked List merupakan solusi yang efisien.

## 6. 
Saat memilih struktur data yang tepat, pemahaman tentang efisiensi operasi fundamental seperti penyisipan (insertion) dan penghapusan (deletion) menjadi krusial. Array dan linked list, dua struktur data linear yang paling umum, menawarkan trade-off yang signifikan dalam hal ini. Secara umum, linked list lebih unggul dalam operasi sisip dan hapus di tengah koleksi data, sementara array dapat lebih efisien untuk operasi di akhir data.

Perbedaan efisiensi ini berakar pada cara kedua struktur data tersebut dialokasikan dalam memori. Array menempati blok memori yang berdekatan (kontigu), sedangkan elemen-elemen dalam linked list (disebut node) tersebar di berbagai lokasi memori dan terhubung melalui pointer.
