```python
product = [
    ["Laptop", 8500000],
    ["Monitor", 2200000],
    ["Tablet", 4500000],
    ["Keyboard", 230000],
    ["Mouse Gaming", 350000],
    ["Mouse", 90000],
    ["Printer", 1500000],
    ["Speaker", 600000],
    ["Headphone", 750000],
    ["Stand Laptop", 50000],
]

def show_product(product):
    print("Daftar Produk dan Harga: ")
    for p in product:
        print(f"{p[0]} : {p[1]:,.0f}")
    print()
    
def bubble_sort_product(product):
    p = len(product)
    for i in range(p):
        for j in range(0, p-i-1):
            if product[j][1]>product[j+1][1]:
                product[j], product[j+1] = product[j+1], product[j]
    return product
    
show_product(product)

bubble_sort_product(product)
show_product(product)

for i in range (1, 6):
    product[-i][1] = int(product[-i][1] * 0.99)

bubble_sort_product(product)
show_product(product)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class Queue:
    def __init__(self, queue_name):
        self.front = None
        self.rear = None
        self.queue_name = queue_name
        
    def enqueue(self, data):
        new_node = Node(data)
        if self.rear is None:
            self.front = self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node
            
    def dequeue(self):
        if self.front is None:
            return None
        temp = self.front
        self.front = temp.next
        if self.front is None:
            self.rear = None
        return temp.data
    
    def is_empty(self):
        return self.front is None
    
    def display(self):
        if self.front is None:
            print(f"Antrian {self.queue_name} kosong")
            return
        print(f"Antrian {self.queue_name}: ", end="")
        temp = self.front
        while temp:
            print(f"[{temp.data}]", end=" -> ")
            temp = temp.next
        print("None")

nasabah = [
    ("Rendra", "Buka Rekening"),
    ("Harun", "Aduan Layanan"),
    ("Maya", "Buka Rekening"),
    ("Ridwan", "Setor Tunai"),
    ("Hina", "Setor Tunai"),
    ("Damar", "Aduan Layanan"),
    ("Joko", "Tarik Tunai"),
    ("Widya", "Aduan Layanan"),
    ("Hana", "Setor Tunai"),
    ("Ahmad", "BUka Rekening")
]

print("Daftar nasabah dan keperluan: ")

for n in nasabah:
    print(f"- {n[0]} : {n[1]}")
    
cs_a = Queue("CS A")
cs_b = Queue("CS B")

for i, n in enumerate(nasabah):
    if i % 2 == 0:
        cs_a.enqueue(f"{n[0]} - {n[1]}")
    else:
        cs_b.enqueue(f"{n[0]} - {n[1]}")

print("\n Antrian setelah nasabah ditambahkan: ")
cs_a.display()
cs_b.display()

print("\n Proses layanan: ")
while not cs_a.is_empty() or not cs_b.is_empty():
    if not cs_a.is_empty():
        served = cs_a.dequeue()
        print(f"{cs_a.queue_name} melayani {served}")
    if not cs_b.is_empty():
        served = cs_b.dequeue()
        print(f"{cs_b.queue_name} melayani {served}")
        
        
        
print("\n=== Semua nasabah sudah dilayani ===")
print("Layanan tutup.")
```

## Singly Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        
    def insert(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        
    def delete(self, key):
        temp = self.head
        
        if temp and temp.data == key:
            self.head = temp.next
            temp = None
            return
        
        prev = None
        while temp and temp.data != key:
            prev = temp
            temp = temp.next
            
        if not temp:
            return
        
        prev.next = temp.next
        temp = None
        
    def traverse(self):
        temp = self.head
        
        while temp:
            print(temp.data, end=' -> ')
            temp = temp.next
        print("None")
        
sll = SinglyLinkedList()
sll.insert(10)
sll.insert(20)
sll.insert(40)
sll.traverse()

sll.delete(20)
sll.traverse()
```

## Doubly Linked List

```python
class DNode:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
        
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        
    def insert(self, data):
        new_node = DNode(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp
        
    def delete(self, key):
        temp = self.head
        while temp and temp.data != key:
            temp = temp.next
        if not temp:
            return
        if temp.prev:
            temp.prev.next = temp.next
        else:
            self.head = temp.next
            
        if temp.next:
            temp.next.prev = temp.prev
            
        temp = None
        
    def traverse_forward(self):
        temp = self.head
        while temp:
            print(temp.data, end=" <-> ")
            last = temp
            temp = temp.next
        print("None")
    
    def traverse_backward(self):
        temp = self.head
        while temp and temp.next:
            temp = temp.next
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.prev
        print("None")

dll = DoublyLinkedList()
dll.insert(10)
dll.insert(20)
dll.insert(30)
dll.traverse_forward()

dll.delete(20)
dll.traverse_forward()
dll.traverse_backward()
```

## Circular Linked List
