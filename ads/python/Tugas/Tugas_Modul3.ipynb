```python
# linear search

def linear_search(arr, x):
    for i in arr:
        if i == x:
            return True
    return False

linear_data = [3, 7, 2, 9, 5, 1, 8, 4, 6, 10, 23, 45, 70, 11, 15]
x = 12

if linear_search(linear_data, x):
    print(f"{x} found in array")
else:
    print(f"{x} not found in arrray")
```

```python
# binary search

def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            return True
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    
    return False

binary_data = [5, 10, 15, 20, 25, 30, 35, 40]
x = 40

if binary_search(binary_data, x):
    print(f"{x} found in array")
else:
    print(f"{x} not found in array")
```

### Sorting

```python
data = [64, 34, 25, 12, 22, 11, 90, 5]

# bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

sort_data = bubble_sort(data)
print("Data bubble sort setelah diurutkan: ", sort_data)

# selection sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

sort_selection = selection_sort(data)
print("Data selection sort setelah diurutkan: ", sort_selection)

# insertion sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

sort_insertion = insertion_sort(data)
print("Data insertion sort setelah diurutkan: ", sort_insertion)

# merge sort
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]
        
        merge_sort(left_half)
        merge_sort(right_half)
        
        i = j = k = 0
        
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1
        
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1
            
        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr

sort_merge = merge_sort(data)
print("Data merge sort setelah diurutkan: ", sort_merge)

# quick sort 
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    
    return quick_sort(left) + middle + quick_sort(right)

sort_quick_sort = quick_sort(data)
print("Data quick sort setelah diurutkan: ", sort_quick_sort)
```

#### Kompleksitas masing masing algoritma
1. Linear search
Kompleksitas algoritma linear search adalah O(n) atau juga disebut kompleksitas linear. Artinya, waktu yang dibutuhkan untuk mencari suatu data berbanding lurus dengan jumlah data yang ada.

##### Kelebihan
- Sangat sederhana dan mudah dipahami
- Tidak perlu menggunakan data yang sudah terurut
- Bekerja dengan baik di kumpulan data kecil
##### Kekurangan
- Sangat lambat untuk data yang besar
- Tidak efisien pada skenario tertentu
- Tidak memanfaatkan data yang sudah terurut

2. Binary Search
Kompleksitas waktu pencarian biner adalah O(log n), baik pada kasus rata-rata maupun kasus terburuk, karena algoritma ini membagi data menjadi dua pada setiap langkah.

##### Kelebihan
- Sangat cepat dan efisien untuk data besar
- Lebih sedikit perbandingan
- Implementasi yang terprediksi
##### Kekurangan 
- Membutuhkan data yang terurut
- Tidak efisien untuk data kecil
- Kurang fleksibel untuk data dinamis

3. Kompleksitas algoritma bubble sort pada kasus rata-rata dan terburuk adalah O(n²) atau kuadratik. Artinya, kinerjanya menurun secara drastis seiring bertambahnya jumlah data. Jika data bertambah 10 kali lipat, waktu pengurutan bisa menjadi 100 kali lebih lama. 

##### Kelebihan 
- Sangat sederhana dan mudah dipahami
- Hemat memori
- Stabil
##### Kekurangan
- Tidak efisien untuk data besar
- Kurang praktis di dunia nyata
- Langkah pengurutan terlalu panjang

4. Kompleksitas algoritma selection sort adalah O(n²) atau kuadratik untuk semua kasus (terbaik, rata-rata, dan terburuk). Ini karena algoritma ini akan selalu melakukan pemindaian penuh untuk mencari elemen terkecil, tidak peduli apakah datanya sudah terurut atau belum.

##### Kelebihan 
- Kinerja konsisten dan tidak terpengaruh urutan data awal
- Sederhana dan hemat memori
- Efektif untuk dataset yang kecil
##### Kekurangan 
- Tidak efisien untuk dataset yang besar
- Tidak Stabil
- Pola akses memori yang tidak berurutan

5. kompleksitas algoritma insertion sort adalah O(n²) atau kuadratik untuk kasus rata-rata dan terburuk (ketika data terurut terbalik). Namun, keunggulannya muncul pada kasus terbaik, di mana kompleksitasnya menjadi O(n) jika data sudah terurut atau hampir terurut. Hal ini membuatnya sangat efisien untuk situasi tersebut.

##### Kelebihan 
- Sederhana dan mudah dipahami
- Cepat untuk data yang hampir terurut
- Stabil dan bekerja secara in-place
##### Kelemahan
- Tidak efisien untuk data yang lebih besar
- Lebih lambat dari bubble sort di skenario terburuk
- Membutuhkan banyak pertukaran

6. kompleksitas algoritma merge sort adalah O(n log n) untuk semua kasus (terbaik, rata-rata, dan terburuk). Ini membuatnya menjadi salah satu algoritma pengurutan berbasis perbandingan yang paling efisien dan dapat diandalkan, terutama untuk data berukuran besar.

##### Kelebihan
- Kompleksitas waktu yang konsisten
- Stabil
- Efisiensi untuk data besar
##### Kekurangan
- Membutuhkan memori tambahan yang besar
- Tidak cocok untuk pengurutan In-Place
- Kurang efisien untuk data kecil

7. Kompleksitas algoritma quick sort pada kasus terbaik dan rata-rata adalah O(n log n), menjadikannya salah satu algoritma pengurutan tercepat dalam praktiknya. Namun, pada kasus terburuk, kinerjanya bisa anjlok menjadi O(n²) atau kuadratik.

##### Kelebihan
- Kecepatan tinggi
- Implementasi yang sederhana
- Efisiensi cache yang baik
##### Kekurangan
- Ketergantungan pada pemilihan pivot
- Tidak stabil
- Tidak ideal untuk data yang kecil
