```python
# linear search

def linear_search(arr, x):
    for i in arr:
        if i == x:
            return True
    return False

linear_data = [3, 7, 2, 9, 5, 1, 8, 4, 6, 10, 23, 45, 70, 11, 15]
x = 12

if linear_search(linear_data, x):
    print(f"{x} found in array")
else:
    print(f"{x} not found in arrray")
```

```python
# binary search

def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            return True
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    
    return False

binary_data = [5, 10, 15, 20, 25, 30, 35, 40]
x = 40

if binary_search(binary_data, x):
    print(f"{x} found in array")
else:
    print(f"{x} not found in array")
```

### Sorting

```python
data = [64, 34, 25, 12, 22, 11, 90, 5]

# bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

sort_data = bubble_sort(data)
print("Data bubble sort setelah diurutkan: ", sort_data)

# selection sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

sort_selection = selection_sort(data)
print("Data selection sort setelah diurutkan: ", sort_selection)

# insertion sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

sort_insertion = insertion_sort(data)
print("Data insertion sort setelah diurutkan: ", sort_insertion)

# merge sort
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]
        
        merge_sort(left_half)
        merge_sort(right_half)
        
        i = j = k = 0
        
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1
        
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1
            
        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr

sort_merge = merge_sort(data)
print("Data merge sort setelah diurutkan: ", sort_merge)

# quick sort 
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    
    return quick_sort(left) + middle + quick_sort(right)

sort_quick_sort = quick_sort(data)
print("Data quick sort setelah diurutkan: ", sort_quick_sort)
```

#### Kompleksitas masing masing algoritma
1. Linear search
Kompleksitas algoritma linear search adalah O(n) atau juga disebut kompleksitas linear. Artinya, waktu yang dibutuhkan untuk mencari suatu data berbanding lurus dengan jumlah data yang ada.

##### Kelebihan
- Sangat sederhana dan mudah dipahami
- Tidak perlu menggunakan data yang sudah terurut
- Bekerja dengan baik di kumpulan data kecil
##### Kekurangan
- Sangat lambat untuk data yang besar
- Tidak efisien pada skenario tertentu
- Tidak memanfaatkan data yang sudah terurut

2. Binary Search
Kompleksitas waktu pencarian biner adalah O(log n), baik pada kasus rata-rata maupun kasus terburuk, karena algoritma ini membagi data menjadi dua pada setiap langkah.

##### Kelebihan
- Sangat cepat dan efisien untuk data besar
- Lebih sedikit perbandingan
- Implementasi yang terprediksi
##### Kekurangan 
- Membutuhkan data yang terurut
- Tidak efisien untuk data kecil
- Kurang fleksibel untuk data dinamis

3. Kompleksitas algoritma bubble sort pada kasus rata-rata dan terburuk adalah O(n²) atau kuadratik. Artinya, kinerjanya menurun secara drastis seiring bertambahnya jumlah data. Jika data bertambah 10 kali lipat, waktu pengurutan bisa menjadi 100 kali lebih lama. 

##### Kelebihan 
- Sangat sederhana dan mudah dipahami
- Hemat memori
- Stabil
##### Kekurangan
- Tidak efisien untuk data besar
- Kurang praktis di dunia nyata
- Langkah pengurutan terlalu panjang

4. Kompleksitas algoritma selection sort adalah O(n²) atau kuadratik untuk semua kasus (terbaik, rata-rata, dan terburuk). Ini karena algoritma ini akan selalu melakukan pemindaian penuh untuk mencari elemen terkecil, tidak peduli apakah datanya sudah terurut atau belum.

##### Kelebihan 
- Kinerja konsisten dan tidak terpengaruh urutan data awal
- Sederhana dan hemat memori
- Efektif untuk dataset yang kecil
##### Kekurangan 
- Tidak efisien untuk dataset yang besar
- Tidak Stabil
- Pola akses memori yang tidak berurutan

5. kompleksitas algoritma insertion sort adalah O(n²) atau kuadratik untuk kasus rata-rata dan terburuk (ketika data terurut terbalik). Namun, keunggulannya muncul pada kasus terbaik, di mana kompleksitasnya menjadi O(n) jika data sudah terurut atau hampir terurut. Hal ini membuatnya sangat efisien untuk situasi tersebut.

##### Kelebihan 
- Sederhana dan mudah dipahami
- Cepat untuk data yang hampir terurut
- Stabil dan bekerja secara in-place
##### Kelemahan
- Tidak efisien untuk data yang lebih besar
- Lebih lambat dari bubble sort di skenario terburuk
- Membutuhkan banyak pertukaran

6. kompleksitas algoritma merge sort adalah O(n log n) untuk semua kasus (terbaik, rata-rata, dan terburuk). Ini membuatnya menjadi salah satu algoritma pengurutan berbasis perbandingan yang paling efisien dan dapat diandalkan, terutama untuk data berukuran besar.

##### Kelebihan
- Kompleksitas waktu yang konsisten
- Stabil
- Efisiensi untuk data besar
##### Kekurangan
- Membutuhkan memori tambahan yang besar
- Tidak cocok untuk pengurutan In-Place
- Kurang efisien untuk data kecil

7. Kompleksitas algoritma quick sort pada kasus terbaik dan rata-rata adalah O(n log n), menjadikannya salah satu algoritma pengurutan tercepat dalam praktiknya. Namun, pada kasus terburuk, kinerjanya bisa anjlok menjadi O(n²) atau kuadratik.

##### Kelebihan
- Kecepatan tinggi
- Implementasi yang sederhana
- Efisiensi cache yang baik
##### Kekurangan
- Ketergantungan pada pemilihan pivot
- Tidak stabil
- Tidak ideal untuk data yang kecil

2. Buat contoh penggunaan masing-masing algoritma dalam sebuah studi kasus

```python
product_data = [
    {'id': 105, 'nama': 'Mouse Gaming', 'harga': 750000},
    {'id': 102, 'nama': 'Keyboard Mechanical', 'harga': 1200000},
    {'id': 101, 'nama': 'Monitor LED 24 inch', 'harga': 1500000},
    {'id': 104, 'nama': 'Webcam HD', 'harga': 450000},
    {'id': 103, 'nama': 'Headset Wireless', 'harga': 950000}
]
```

```python
# Linear search - Mencari produk berdasarkan nama
def linear_search_by_name(data, target_name):
    for item in data:
        if item['nama'] == target_name:
            return item
    return None

search_item = "Mouse Gaming"
result = linear_search_by_name(product_data, search_item)

if result:
    print(f"Linear search: Produk '{search_item}' ditemukan -> {result}.")
else:
    print(f"Linear search: Produk '{search_item}' tidak ditemukan .")
```

```python
# Binary search - Mencari produk berdasarkan 'id' yang telah diurutkan
sorted_id = sorted(product_data, key=lambda x: x['id'])

def binary_search_by_id(data, target_id):
    low = 0
    high = len(data) - 1
    
    while low <= high:
        mid  = (low + high) // 2
        if data[mid]['id'] == target_id:
            return data[mid]
        elif target_id < data[mid]['id']:
            high = mid - 1
        else:
            low = mid + 1
    return None

search_id = 104
print("\nData untuk binary search (sort by ID): ")
for p in sorted_id:
    print(p)
    
result = binary_search_by_id(sorted_id, search_id)

if result:
    print(f"\nBinary search: Produk dengan ID {search_id} ditemukan -> {result}.")
else:
    print(f"\nBinary search: Produk dengan ID {search_id} tidak ditemukan.")
```

```python
# Bubble sort - Mengurutkan barang berdasarkan harga barang
def bubble_sort_by_price(data):
    n = len(data)
    
    arr = data[:]
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j]['harga'] > arr[j + 1]['harga']:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

print("--- Bubble sort (Berdasarkan harga) ---")
print("\nSebelum: ")
for p in product_data:
    print(p)
sorted_product_bubble = bubble_sort_by_price(product_data)
print("\nSesudah: ")
for bs in sorted_product_bubble:
    print(bs)
```

```python
# Selection sort - Mengurutkan barang berdasarkan harga

def selection_sort_by_price(data):
    n = len(data)
    arr = data[:]
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j]['harga'] < arr[min_idx]['harga']:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

print("\n--- Selection Sort (berdasarkan harga) ---")
print("\nSebelum:")
for p in product_data:
    print(p)
product_sorted_selection = selection_sort_by_price(product_data)
print("\nSesudah:")
for ss in product_sorted_selection:
    print(ss)
```

```python
# Insertion sort - Mengurutkan barang berdasarkan harga

def insertion_sort_by_price(data):
    arr = data[:]
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key['harga'] < arr[j]['harga']:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr

print("\n--- Insertion Sort (berdasarkan harga) ---")
print("\nSebelum:")
for p in product_data:
    print(p)
product_sorted_insertion = insertion_sort_by_price(product_data)
print("\nSesudah:")
for si in product_sorted_insertion:
    print(si)
```

```python
# Merge sort - Mengurutkan barang berdasarkan harga

def merge_sort_by_price(data):
    arr = data[:]
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort_by_price(L) 
        merge_sort_by_price(R)

        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i]['harga'] < R[j]['harga']:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
        
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr

print("\n--- Merge Sort (berdasarkan harga) ---")
print("Sebelum:")
for p in product_data:
    print(p)
product_sorted_merge = merge_sort_by_price(product_data)
print("Sesudah:")
for sm in product_sorted_merge:
    print(sm)
```

```python
# Quick sort - Mengurutkan barang berdasarkan harga

def quick_sort_by_price(data):
    arr = data[:]
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr) // 2]['harga']
        left = [x for x in arr if x['harga'] < pivot]
        middle = [x for x in arr if x['harga'] == pivot]
        right = [x for x in arr if x['harga'] > pivot]
        return quick_sort_by_price(left) + middle + quick_sort_by_price(right)

print("\n--- Quick Sort (berdasarkan harga) ---")
print("\nSebelum:")
for p in product_data:
    print(p)
product_sorted_quick = quick_sort_by_price(product_data)
print("\nSesudah:")
for sq in product_sorted_quick:
    print(sq)
```

```python
import matplotlib.pyplot as plt
import numpy as np

n = 1000
algorithms = [
    'Bubble sort', 
    'Selection sort', 
    'Insertion sort', 
    'Merge sort', 
    'Quick sort'
]

best_case = {
    'Bubble sort' : n,
    'Selection sort' : n**2,
    'Insertion sort' : n,
    'Merge sort' : n * np.log2(n),
    'Quick sort' : n * np.log2(n)
}

average_case = {
    'Bubble sort' : n**2,
    'Selection sort' : n**2,
    'Insertion sort' : n**2,
    'Merge sort' : n * np.log2(n),
    'Quick sort' : n * np.log2(n)
}

worst_case = {
    'Bubble sort' : n**2,
    'Selection sort' : n**2,
    'Insertion sort' : n**2,
    'Merge sort' : n * np.log2(n),
    'Quick sort' : n**2
}

best = [best_case[algo] for algo in algorithms]
average = [average_case[algo] for algo in algorithms]
worst = [worst_case[algo] for algo in algorithms]

x = np.arange(len(algorithms))
width = 0.25

fig, ax = plt.subplots(figsize = (12, 7))
rects1 = ax.bar(x - width, best, width, label='Best Case', color='skyblue')
rects2 = ax.bar(x, average, width, label='Average Case', color='salmon')
rects3 = ax.bar(x + width, worst, width, label='Worst Case', color='lightgreen')

ax.set_ylabel('Estimasi Jumlah Operasi (Skala Log)')
ax.set_title('Perbandingan Kompleksitas Waktu Algoritma Pengurutan (n = 1000)')
ax.set_xticks(x)
ax.set_xticklabels(algorithms)
ax.set_yscale('log')
ax.legend()

ax.bar_label(rects1, padding=3, fmt='%.0e')
ax.bar_label(rects2, padding=3, fmt='%.0e')
ax.bar_label(rects3, padding=3, fmt='%.0e')

fig.tight_layout()
plt.show()
```
